## Load Data for n = [101,1001,2001,4001,8001]

```python
import numpy as np
steps = [101,1001,2001,4001,8001]
data = ["data_101","data_1001","data_2001","data_4001","data_8001"]
mapp = {}

for n,d in zip(steps,data):
    
    d = np.load(f"solutions_{n}.npz", allow_pickle=True)
    mapp[n] = d



print(mapp)
```

```python
import matplotlib.pyplot as plt 

for n,data in mapp.items():
    

    x = np.linspace(0,1,n)
    phi_solution = data["phi_solution"]
    plt.figure(figsize=(7,5))
    
    plt.plot(x,phi_solution[0][0],'k--', label='phi_numerical')
    plt.plot(x,phi_solution[0][1], 'r--', label='phi_exact')
    plt.title(f"phi(x) solution at T=0.01, n={n}")
    plt.grid(True)
    plt.show()
```
